# Important Things About Python
* dynamically typed
  1. we do not specify the type of data that will go into our variables
     *  just put a value in a variable
  2. you can later overwrite the variable with a different type of data
* everything is an object
  * every "thing" in Python consists of multiple parts that we can inspect
* built-in functions do not change the data ("arguments") that are passed into them
  * if you want to change an object in Python, you must call/invoke/apply a method on/to it
    * not all methods change the object they are called/invoked/applied on/to
  * __`sorted(container)`__
    1. leaves the original container alone and returns a new sorted list
    2. argument can be any container, not just a list
* scalar vs. container
  * scalar = "single value"
    * e.g., bool, float, int
  * container = 0+ values
    * e.g., str, list
* mutable vs. immutable objects
  * mutable = changeable (e.g., list, dict)
  * immutable = not changeable (e.g., str, tuple)

# "Pythonic"
* using idioms and ways of writing Python that other Python programmers expect to see
  * "My name is Rick and I'm a Java programmer. I've been teaching myself Python, but my Python looks like Java."
* __`container[-1]`__ always means last item in the container
* __`container[:n]`__ always means the first n items in the container
* __`container[n:]`__ always means the rest of the above, i.e., from n to the end
* __`container[-n:]`__ always means the last n items in the container
* __`for _ in range(n)`__ means "do this loop n times"
* if a variable/datatype doesn't do what we want, consider changing its type
  * convert ints to strings in order to index and use str methods
  * strings are immutable, so consider turning them into lists, manipulating them as lists, and then converting to strings

## Quick Lab: String Functions
* write a Python program to read in a sentence and tell the user how many vowels are in that sentence
* so if the user entered "Apples are my favorite fruit", your program would respond with 10 (or 11 if you count 'y' as a vowel)
* output the original string with any vowels "highlighted" by making them upper case, e.g., **ApplEs ArE my fAvOrItE frUIt**

```python
# What do humans do to solve this problem?
# 1. scan or look at letters left to right
# 2. if you see a, e, i, o, u (y), then count it
# 3. tell us the result

# Pseudocode
# 0. get text from the user (kind of like the whiteboard)
# 0a. set count to 0
# 1. look at each letter / for each letter
# 2.   if it's a vowel 
# 3.      add 1 to the count (increment)
# 4. output/print the count

text = input('Enter some text and I will count the vowels: ') # 0
vowel_count = 0 # 0a

for character in text.lower(): # 1
    if character in 'aeiouy': # 2
        vowel_count += 1 # 3

print(vowel_count) # 4
```

```python
'b' in 'aeiouy'
```

```python
# What do humans do to solve this problem?
# 1. look at the sentence
# 2. write the sentence again, where...
# 3. each vowel is written as upper case
# 4. each consonant is written as lower case

# Pseudocode
# 1. for each character in the sentence:
# 2.    if it's a vowel:
# 3.       write it as upper case
# 4.    else:
# 5.       write it as lower case
# 6. go to next line

for character in text.lower(): # 1
    if character in 'aeiouy': # 2
        print(character.upper(), end='') # 3
    else: # 4
        print(character.lower(), end='') # 5

print() # 6 
```

## Quick Lab: Lists
* Write a program that asks the user to input two lists and then finds and prints the common elements between them
<pre>
Enter a list of items: <b>apple cherry banana lemon</b>
Enter a second list of items: <b>apple guava banana lime</b>
Common elements: apple banana

```python
# 1. get first list of items
# 2. get second list of items
# 2a. create a comm
# 3. for each item in first list:
# 4.    if item is in second list:
# 5.       print item (or add to a new list)
```

```python
list1 = input('Enter a list of items: ').split() # 1
list2 = input('Enter a list of items: ').split() # 2
common = [] # 2a

for item in list1: # 3
    if item in list2: # 4
        common.append(item) # 5 (or print)

print(common)
```

```python
print(', '.join(common))
```

```python
print(common)
```

```python
print(common[0], common[1])
```

```python
print(*common) # * = "unpack" operator, and it does the equivalent of the above
```
