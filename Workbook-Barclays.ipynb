# Important Things About Python
* dynamically typed
  1. we do not specify the type of data that will go into our variables
     *  just put a value in a variable
  2. you can later overwrite the variable with a different type of data
* everything is an object
  * every "thing" in Python consists of multiple parts that we can inspect
* built-in functions do not change the data ("arguments") that are passed into them
  * if you want to change an object in Python, you must call/invoke/apply a method on/to it
    * not all methods change the object they are called/invoked/applied on/to
  * __`sorted(container)`__
    1. leaves the original container alone and returns a new sorted list
    2. argument can be any container, not just a list
* scalar vs. container
  * scalar = "single value"
    * e.g., bool, float, int
  * container = 0+ values
    * e.g., str, list, dict, set
* mutable vs. immutable objects
  * mutable = changeable (e.g., list, dict, set)
  * immutable = not changeable (e.g., str, tuple)
* ordered vs. unordered datatypes
  * ordered = str, list
  * unordered = dict, set

# "Pythonic"
* using idioms and ways of writing Python that other Python programmers expect to see
  * "My name is Rick and I'm a Java programmer. I've been teaching myself Python, but my Python looks like Java."
* __`container[-1]`__ always means last item in the container
* __`container[:n]`__ always means the first n items in the container
* __`container[n:]`__ always means the rest of the above, i.e., from n to the end
* __`container[-n:]`__ always means the last n items in the container
* __`for _ in range(n)`__ means "do this loop n times"
* if a variable/datatype doesn't do what we want, consider changing its type
  * convert ints to strings in order to index and use str methods
  * strings are immutable, so consider turning them into lists, manipulating them as lists, and then converting to strings
* composing functions, e.g., __`input('Enter something: ').lower().split()`__
* Python "truthiness"
  * 0 and 0.0 are considered False in a Boolean context
    * non-zero numbers (including negative) are considered True in a Boolean context
  * empty containers (e.g., '', [], {}) are considered False in a Boolean context
    * non-empty contains (e.g., 'string', [1], {1: 'one'}) are considered True in Boolean context
  * __`None`__ is considered False in a Boolean context

## Quick Lab: String Functions
* write a Python program to read in a sentence and tell the user how many vowels are in that sentence
* so if the user entered "Apples are my favorite fruit", your program would respond with 10 (or 11 if you count 'y' as a vowel)
* output the original string with any vowels "highlighted" by making them upper case, e.g., **ApplEs ArE my fAvOrItE frUIt**

```python
# What do humans do to solve this problem?
# 1. scan or look at letters left to right
# 2. if you see a, e, i, o, u (y), then count it
# 3. tell us the result

# Pseudocode
# 0. get text from the user (kind of like the whiteboard)
# 0a. set count to 0
# 1. look at each letter / for each letter
# 2.   if it's a vowel 
# 3.      add 1 to the count (increment)
# 4. output/print the count

text = input('Enter some text and I will count the vowels: ') # 0
vowel_count = 0 # 0a

for character in text.lower(): # 1
    if character in 'aeiouy': # 2
        vowel_count += 1 # 3

print(vowel_count) # 4
```

```python
'b' in 'aeiouy'
```

```python
# What do humans do to solve this problem?
# 1. look at the sentence
# 2. write the sentence again, where...
# 3. each vowel is written as upper case
# 4. each consonant is written as lower case

# Pseudocode
# 1. for each character in the sentence:
# 2.    if it's a vowel:
# 3.       write it as upper case
# 4.    else:
# 5.       write it as lower case
# 6. go to next line

for character in text.lower(): # 1
    if character in 'aeiouy': # 2
        print(character.upper(), end='') # 3
    else: # 4
        print(character.lower(), end='') # 5

print() # 6 
```

## Quick Lab: Grep Knockoff
* __`grep`__ is a Linux program that finds all lines in file that contain a certain pattern, e.g., __`grep Python file.txt`__ will print out any lines of the file file.txt that have the word __`Python`__ in the line
* write a version of __`grep`__ in Python
  * your program should ask the user for two things:
    * a filename
    * a pattern to search for
  * ...and then will print out all lines containing the pattern, e.g.,

<pre><b>
Enter a file name: poem.txt
Enter pattern to search for: wood
</b></pre>

```python
# 1. get name of file
# 2. get pattern/chars to search for
# 3. for each line in the file:
# 4.    if the pattern is found in that line
# 5.       print the line
```

```python
filename = input('Enter a filename: ') # 1
pattern = input('Enter the pattern to search for: ') # 2

with open(filename) as file: # add a with block to be good Python folks
    for line in file: # 3
        if pattern.lower() in line.lower(): # 4
            print(line, end='') # 5
```

## Group Lab/Homework: File I/O + dicts
* write a Python program to read a file and count the number of occurrences of each word in the file
* use a __`dict`__, indexed by word, to count the occurrences
* remember __`d.get(key)`__ will return __`None`__ if there is no such key in the dict (vs. __`d[key]`__ which will throw an exception) and also the __`in`__ operator
* treat __The__ and __the__ as the same word when counting
* print out words and counts, from most common to least common
* __EXTRA:__ remove punctuation, so __Hamlet,__ == __Hamlet__ # refer back to "import this"
* The Road Not Taken and Hamlet are in your materials

```python
# 0. start w/an empty dict to hold words and counts
# 1. get the name of a file
# 2. open the file
# 3. for each line of the file:
# 4.    split into words
# 5.    for each word in the line:
# 6.       if we've seen the word before:
# 7.          incremement count
# 8.       else:
# 9.          make a new entry for that word and set count to 1
# 10. print out the dict
```

```python
wordcount = {} # 0
filename = input('Enter filename: ') # 1

for line in open(filename): # 3
    for word in line.lower().split(): # 4, 5
        if word in wordcount: # 6 (if there's an entry for it)
            wordcount[word] += 1 # 7
        else: # 8
            wordcount[word] = 1 # 9        
```

```python
for key in sorted(wordcount):
    print(key, wordcount[key])
```

```python
for key in sorted(wordcount, key=wordcount.get, reverse=True)[:10]:
    print(key, wordcount[key])
```

```python
def calculate(num1, num2, operator):
    """Take two numbers and an operator and compute and return result"""
    if operator == '+':
        return num1 + num2
    if operator == '-':
        return num1 - num2
    if operator == '*':
        return num1 * num2
    if operator == '/': # floating point div
        return num1 / num2
    if operator == '//': # int div
        return num1 // num2
    print('bad operator:', operator)
    print('valid operators are +, -, *, /, //')
```

```python
calculate(2, 4, '+')
```

```python
import string
string.ascii_lowercase
```

```python
# 1. create set of the input letters (we have to filter out non-letters)
# 2. create set of the alphabet
# 3. they should be equal
```

```python
# same as above but don't need the alphabet
# 1. create set of the input letters
# 2. it's a pangram if there are 26
```

```python
# 1. create set of the alphabet
# 2. look at each input letter
# 3. remove it from the set
# 4. how many should be left? 0
```

```python
def is_pangram(text):
    """Return True if text is a pangram, False otherwise"""
    # use a list comprehension to *filter* out non-letters from the input
    # (that is, only keep letters, discard everything else)
    # and then set-ify to remove duplicates
    input_letters = set([letter for letter in text.lower()
                                if letter in string.ascii_lowercase])
    # there should be 26...
    return len(input_letters) == 26
```

```python
is_pangram('# The quick brown fox jumps over the lazy dog')   
```

```python
def is_pangram2(text):
    """Return True if text is a pangram, False otherwise"""
    # for each letter in input, remove it from a set of the letters
    # if we end up with 0 remaining, it's a pangram
    alphabet = set(string.ascii_lowercase)
    for character in text.lower():
        alphabet.discard(character) # discard so no error if not there
    return len(alphabet) == 0
```

```python
is_pangram2('# The quick brown fox jumps over the lazy dog')   
```

```python
def is_pangram3(text):
    """Return True if text is a pangram, False otherwise"""
    unique_letters_found = set() # start w/empty set
    for character in text.lower():
        if character in string.ascii_lowercase: # is it a-z?
            unique_letters_found.add(character)

    return len(unique_letters_found) == 26
```

```python
is_pangram3('# The wizard quickly jinxed the gnomes before they vaporized')
```
